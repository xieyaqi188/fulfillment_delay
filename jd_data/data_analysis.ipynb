```python
import pandas as pd
import numpy as np
```

```python
order = pd.read_csv("./JD_data/JD_order_data.csv")
order.info()
```

```python
# 1P
order_1p = order[order['type'] == 1]
order_1p.info()
order_id_list = order_1p['order_ID'].unique().tolist()
```

```python
dc_list = order['dc_des'].unique().tolist()
print(dc_list, len(dc_list))
```

```python
for fdc in dc_list:
    order_fdc_1p = order_1p[order_1p['dc_des'] == fdc]
    # order number
    order_fdc_id_list = order_fdc_1p['order_ID'].unique().tolist()
    # number of single-item and multi-item
    order_id_s = 0
    order_id_m = 0
    for i in order_fdc_id_list:
        tmp = order_fdc_1p[order_fdc_1p['order_ID'] == i]
        tmp1 = tmp['quantity'].sum()
        # sku1 * 1
        if len(tmp) == 1 and tmp1 == 1:
                order_id_s += 1
        else:
            order_id_m += 1
    print("fdc", fdc, len(order_fdc_id_list), order_id_s, order_id_m, order_id_s / (order_id_s + order_id_m))
```

```python
tt = 0
for i in order_id_list:
    tmp = order_1p[order_1p['order_ID'] == i]
    tmp1 = tmp['dc_ori'].unique().tolist()
    if len(tmp1) > 1:
       print("a")
       tt += 1
print(tt)
```

## Specific destination center

```python
# FDC 5 as destination center
order_des_dc5 = order[order['dc_des'] == 5]
order_des_dc5 = order_des_dc5.drop_duplicates(subset=['order_ID', 'sku_ID', 'order_time'])
# Destination FDC 5 + 1P orders
order_des_dc5_1p = order_des_dc5[order_des_dc5['type'] == 1]
# order number
order_id_list = order_des_dc5_1p['order_ID'].unique().tolist()
print(len(order_id_list))
```

```python
# number of order types
order_id_s_s = []
order_id_s_m = []
order_id_m_s = []
order_id_m_m = []
for i in order_id_list:
    tmp = order_des_dc5_1p[order_des_dc5_1p['order_ID'] == i]
    tmp1 = tmp[tmp['quantity'] == 1]
    # sku1 * 1
    if len(tmp) == 1 and len(tmp1) == 1:
        order_id_s_s.append(i)
    # sku1 * n
    if len(tmp) == 1 and len(tmp1) == 0:
        order_id_s_m.append(i)
    # sku1 * 1 + sku2 * 1
    if 1 < len(tmp) == len(tmp1):
        order_id_m_s.append(i)
    # sku1 * 1 + sku2 * n
    if len(tmp) > 1 and len(tmp1) < len(tmp):
        order_id_m_m.append(i)
print(len(order_id_s_s), len(order_id_s_m), len(order_id_m_s), len(order_id_m_m), len(order_id_list))
print(len(order_id_s_s)/ len(order_id_list))
```

```python
tt = 0
for i in order_id_list:
    tmp = order_des_dc5_1p[order_des_dc5_1p['order_ID'] == i]
    tmp1 = tmp['dc_ori'].unique().tolist()
    if len(tmp1) > 1:
       print("a")
       tt += 1
print(tt)
```

```python
# s-s: sku1 * 1
order_s_s = order_des_dc5_1p[order_des_dc5_1p['order_ID'].isin(order_id_s_s)]
sku_s_s = order_s_s['sku_ID'].unique().tolist()
print(len(sku_s_s))
```

```python
# m-s: sku1 * 1 + sku2 * 1
order_m_s = order_des_dc5_1p[order_des_dc5_1p['order_ID'].isin(order_id_m_s)]
sku_m_s = order_m_s['sku_ID'].unique().tolist()
print(len(sku_m_s))
```

```python
# single-single orders involving multi-single skus
order_s_s_sku = order_s_s[order_s_s['sku_ID'].isin(sku_m_s)]
print(len(order_s_s_sku))
```

```python
demand_m_s = {}
inventory_m_s = {}
# demand for single-single orders
for i in sku_m_s:
    tmp = order_s_s[order_s_s['sku_ID'] == i]
    tmp_index = str(sku_m_s.index(i))
    demand_m_s[tmp_index] = len(tmp)
    inventory_m_s[tmp_index] = len(tmp)
print(demand_m_s)
```

```python
# demand for multi-single orders
for i in order_id_m_s:
    tmp = order_des_dc5_1p[order_des_dc5_1p['order_ID'] == i]
    tmp_list = tmp['sku_ID'].values.tolist()
    tmp_list2 = [sku_m_s.index(j) for j in tmp_list]
    tmp_list3 = sorted(tmp_list2)
    tmp_list3 = list(dict.fromkeys(tmp_list3))
    tmp_key = str(tmp_list3[0])
    inventory_m_s[tmp_key] += 1
    for k in range(1, len(tmp_list3)):
        tmp_key = tmp_key + "-" + str(tmp_list3[k])
        inventory_m_s[str(tmp_list3[k])] += 1
    if tmp_key in demand_m_s:
        demand_m_s[tmp_key] += 1
    else:
        demand_m_s[tmp_key] = 1
print(demand_m_s)
```

## network: RDC-FDC

```python
network = pd.read_csv("./JD_data/JD_network_data.csv")
network.info()
```

```python
network['region_ID'].value_counts()
```

```python
# RDC
rdc_list = fdc_list = network['region_ID'].unique().tolist()
print(rdc_list)
```

```python
# RDC orders
rdc_order_num = {}
for rdc in rdc_list:
    tmp_rdc_order = order[order['dc_des'] == rdc]
    tmp_rdc_order = tmp_rdc_order.drop_duplicates(subset=['order_ID', 'sku_ID', 'order_time'])
    tmp_rdc_order_1p = tmp_rdc_order[tmp_rdc_order['type'] == 1]
    rdc_order_num[rdc] = len(tmp_rdc_order_1p)
print(rdc_order_num)
```

```python
# FDC
fdc_list = network['dc_ID'].unique().tolist()
print(fdc_list)
```

```python
rdc = int(network[network['dc_ID'] == 26]['region_ID'])
print(rdc)
```

```python
# FDC orders
dc_order = pd.DataFrame(columns=['DC', 'RDC', 'total_order_num', 'order_num_DC', 'order_num_RDC', 'sum_DC_RDC'], dtype='Int64')
for fdc in fdc_list:
    rdc = int(network[network['dc_ID'] == fdc]['region_ID'])
    tmp = []
    tmp_fdc_order = order[order['dc_des'] == fdc]
    tmp_fdc_order = tmp_fdc_order.drop_duplicates(subset=['order_ID', 'sku_ID', 'order_time'])
    tmp_fdc_order_1p = tmp_fdc_order[tmp_fdc_order['type'] == 1]
    tmp1 = tmp_fdc_order_1p[tmp_fdc_order_1p['dc_ori'] == fdc]
    tmp2 = tmp_fdc_order_1p[tmp_fdc_order_1p['dc_ori'] == rdc]
    if fdc == rdc:
        tmp = [fdc, rdc, rdc_order_num[fdc], 0, len(tmp1), len(tmp1)]
    else:
        tmp = [fdc, rdc, len(tmp_fdc_order_1p), len(tmp1), len(tmp2), len(tmp1) + len(tmp2)]
    dc_order.loc[len(dc_order.index)] = tmp
print(dc_order)
dc_order.to_csv('./pre_data/dc_order.csv', index=False)
```
